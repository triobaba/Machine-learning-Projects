```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib_inline
```

```python
p=pd.read_csv("train.csv")
```

```python
plt.scatter(p.area,p.price)
plt.xlabel("AREA")
plt.ylabel("PRICE")
plt.title("prediction model of price vs Area")
```

```python
reg=linear_model.LinearRegression()
reg.fit(p[["area"]],p.price)
```

```python
reg.predict=int(input("enter the area of the house to be predicted"))
```
